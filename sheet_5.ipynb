## Question 1
Write a NumPy program that creates a NumPy array of random numbers and uses SciPy to compute the statistical properties (mean, median, variance) of the array.

```python
import numpy as np
from scipy import stats

# Create a NumPy array of random numbers
random_array = np.random.rand(1000)

# Compute statistical properties
mean = np.mean(random_array)
median = np.median(random_array)
variance = np.var(random_array)

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Variance: {variance}")
```

## Question 2
Write a NumPy program to generate a 2D array and performs a discrete Fourier transform using SciPy's fftpack module.

```python
import numpy as np
from scipy.fftpack import fft2, fftshift
import matplotlib.pyplot as plt

# Generate a simple 2D array
x = np.linspace(0, 10, 100)
y = np.linspace(0, 10, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X)  # Simple sine function along X-axis

# Perform 2D Fourier transform
Z_fft = fftshift(fft2(Z))
# Print the input function
print("Input function:")
print(Z)

# Print the Fourier transformation
print("\nFourier transformation:")
print(np.abs(Z_fft))


```

## Question 3
Write a NumPy program that creates a NumPy array and uses SciPy to solve a linear algebra problem. For example, finding the determinant, inverse, and eigenvalues of a matrix.

```python
import numpy as np
from scipy import linalg

# Create a NumPy array (matrix)
A = np.array([[1, 2], [3, 4]])

# Find determinant
det = linalg.det(A)
print(f"Determinant: {det}")

# Find inverse
inv = linalg.inv(A)
print(f"Inverse:\n{inv}")

# Find eigenvalues and eigenvectors
eigenvalues, eigenvectors = linalg.eig(A)
print(f"Eigenvalues: {eigenvalues}")
print(f"Eigenvectors:\n{eigenvectors}")
```

## Question 4
Write a NumPy program to generate a set of data points using NumPy and perform interpolation using SciPy's interpolate module.

```python
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt

# Generate data points
x = np.linspace(0, 10, 10)
y = np.sin(x)

# Create interpolation function
f = interpolate.interp1d(x, y, kind='cubic')

# Generate points for smooth curve
x_new = np.linspace(0, 10, 100)
y_new = f(x_new)

# Plot original and interpolated data
plt.figure(figsize=(10, 6))
plt.scatter(x, y, label='Original data')
plt.plot(x_new, y_new, 'r', label='Interpolated data')
plt.legend()
plt.title('Cubic Interpolation')
plt.show()
```

## Question 5
Write a NumPy program to create a time series dataset and apply SciPy's signal processing functions to filter the data.

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

# Generate time series data
t = np.linspace(0, 1, 1000, endpoint=False)
data = np.sin(2*np.pi*10*t) + 0.5*np.sin(2*np.pi*20*t) + np.random.normal(0, 0.1, 1000)

# Apply low-pass filter
b, a = signal.butter(4, 0.1, 'low', analog=False)
filtered_data = signal.filtfilt(b, a, data)

# Plot original and filtered data
plt.figure(figsize=(12, 6))
plt.plot(t, data, label='Original signal')
plt.plot(t, filtered_data, 'r', label='Filtered signal')
plt.legend()
plt.title('Low-pass Filter Applied to Time Series Data')
plt.show()
```

## Question 6
You will simulate sales data for a small retail store for 12 months and for 4 products, analyze the data, and perform basic statistical operations like total sales, average sales, maximum sales, etc. You will also determine the best and worst performing months based on the sales data.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sales data
np.random.seed(42)
products = ['Product A', 'Product B', 'Product C', 'Product D']
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales_data = np.random.randint(100, 1000, size=(4, 12))

# Calculate total and average sales
total_sales = np.sum(sales_data)
avg_sales = np.mean(sales_data)
max_sales = np.max(sales_data)

print(f"Total sales: ${total_sales}")
print(f"Average monthly sales: ${avg_sales:.2f}")
print(f"Maximum monthly sales: ${max_sales}")

# Determine best and worst performing months
monthly_totals = np.sum(sales_data, axis=0)
best_month = months[np.argmax(monthly_totals)]
worst_month = months[np.argmin(monthly_totals)]

print(f"Best performing month: {best_month}")
print(f"Worst performing month: {worst_month}")

# Plot sales data
plt.figure(figsize=(12, 6))
for i, product in enumerate(products):
    plt.plot(months, sales_data[i], label=product)
plt.legend()
plt.title('Monthly Sales by Product')
plt.xlabel('Month')
plt.ylabel('Sales ($)')
plt.show()
```

## Question 7
This project involves analysing the marks of students across multiple subjects. The goal is to generate and analyse data, including calculating total marks, average marks, subject-wise performance, and identifying top and bottom performers. You will also determine the passing percentage and generate insights based on students' performance.

Student Name | Math | Physics | Chemistry | English
-------------|------|---------|-----------|--------
Arin         | 85   | 78      | 92        | 88
Aditya       | 79   | 82      | 74        | 90
Chirag       | 90   | 85      | 89        | 92
Gurleen      | 66   | 75      | 80        | 78
Kunal        | 70   | 68      | 75        | 85

```python
import numpy as np
import matplotlib.pyplot as plt

# Create arrays for student data
student_names = np.array(['Arin', 'Aditya', 'Chirag', 'Gurleen', 'Kunal'])
subjects = np.array(['Math', 'Physics', 'Chemistry', 'English'])
marks = np.array([
    [85, 78, 92, 88],
    [79, 82, 74, 90],
    [90, 85, 89, 92],
    [66, 75, 80, 78],
    [70, 68, 75, 85]
])

# Calculate total and average marks
total_marks = np.sum(marks, axis=1)
average_marks = np.mean(marks, axis=1)

# Identify top and bottom performers
top_performer = student_names[np.argmax(total_marks)]
bottom_performer = student_names[np.argmin(total_marks)]

# Print student data
print("Student Data:")
for i, student in enumerate(student_names):
    print(f"{student}: {marks[i]} Total: {total_marks[i]} Average: {average_marks[i]:.2f}")

print(f"\nTop performer: {top_performer}")
print(f"Bottom performer: {bottom_performer}")

# Calculate passing percentage (assuming passing mark is 60)
passing_percentage = np.mean(np.all(marks >= 60, axis=1)) * 100
print(f"Passing percentage: {passing_percentage:.2f}%")

# Subject-wise performance
subject_averages = np.mean(marks, axis=0)
print("\nSubject-wise average performance:")
for subject, avg in zip(subjects, subject_averages):
    print(f"{subject}: {avg:.2f}")

# Visualize subject-wise performance
plt.figure(figsize=(10, 6))
plt.bar(subjects, subject_averages)
plt.title('Subject-wise Average Performance')
plt.ylabel('Average Score')
plt.show()
```

## Question 8
You are provided with the following data for the velocity of an object over time. The velocity data is modelled by a quadratic function of the form v(t)=at^2+bt+c, where v(t) is the velocity at time t, and a, b, and c are constants to be determined using curve fitting. (Using SciPy's Curve fitting)

Time (Seconds) | 0 | 1   | 2   | 3    | 4    | 5
---------------|---|-----|-----|------|------|------
Velocity (m/s) | 2 | 3.1 | 7.9 | 18.2 | 34.3 | 56.2

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Define the quadratic function
def quadratic(t, a, b, c):
    return a * t**2 + b * t + c

# Data
time = np.array([0, 1, 2, 3, 4, 5])
velocity = np.array([2, 3.1, 7.9, 18.2, 34.3, 56.2])

# Perform curve fitting
popt, _ = curve_fit(quadratic, time, velocity)

# Extract the fitted parameters
a, b, c = popt

# Generate points for the fitted curve
t_fit = np.linspace(0, 5, 100)
v_fit = quadratic(t_fit, a, b, c)

# Plot the results
plt.figure(figsize=(10, 6))
plt.scatter(time, velocity, label='Data')
plt.plot(t_fit, v_fit, 'r-', label='Fitted curve')
plt.xlabel('Time (seconds)')
plt.ylabel('Velocity (m/s)')
plt.title('Velocity vs Time: Quadratic Fit')
plt.legend()
plt.grid(True)
plt.show()

print(f"Fitted equation: v(t) = {a:.2f}t^2 + {b:.2f}t + {c:.2f}")
```
